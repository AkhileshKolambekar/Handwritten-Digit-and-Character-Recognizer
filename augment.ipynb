```python
import glob

import cv2

import matplotlib.pyplot as plt

from PIL import Image, ImageOps

import numpy as np
```

```python
def zoom_out(charac_path,img_path):

    img_name = img_path[img_path.rindex('/')+1:]
    final_path = charac_path+"/zoom_out_1_"+img_name

    image = Image.open(img_path)

    new_width = int(image.width / 2) #zoom factor
    new_height = int(image.height / 2)
    
    resized_image = image.resize((new_width, new_height), Image.LANCZOS)
    
    new_image = Image.new("RGB", (image.width, image.height), (255, 255, 255))

    top_left_x = (image.width - new_width) // 2
    top_left_y = (image.height - new_height) // 2

    new_image.paste(resized_image, (top_left_x, top_left_y))

    new_image.save(final_path)
```

```python
# Zoom out the images by zooming_factor = 2
for i in glob.glob('D:/Dataset/new_trying/Train/*'):
    charac_path = i.replace('\\','/')
    for j in glob.glob(charac_path+"/*"):
        img_path = j.replace('\\','/')
        zoom_out(charac_path,img_path)
    print(charac_path+" completed")
```

```python
# Zoom out the images by zooming_factor = 2
for i in glob.glob('D:/Dataset/new_trying/Test/*'):
    charac_path = i.replace('\\','/')
    for j in glob.glob(charac_path+"/*"):
        img_path = j.replace('\\','/')
        zoom_out(charac_path,img_path)
    print(charac_path+" completed")
```

```python
def rotate_image(charac_path,img_path):

    img_name = img_path[img_path.rindex('/')+1:]

    img = Image.open(img_path)
    white = (255,255,255)

    rot_left = img.rotate(20, Image.NEAREST ,expand = 1, fillcolor=white)
    rot_left.save(charac_path+'/rot_left_'+img_name)

    rot_right = img.rotate(-20, Image.NEAREST ,expand = 1, fillcolor=white)
    rot_right.save(charac_path+'/rot_right_'+img_name)
```

```python
for i in glob.glob('D:/Dataset/new_trying/Train/*'):
    charac_path = i.replace('\\','/')
    for j in glob.glob(charac_path+"/*"):
        img_path = j.replace('\\','/')
        rotate_image(charac_path,img_path)
    print(charac_path+" completed")
```
